```bash

```

수업 자료에서는 대부분 루트 폴더에서 명령을 실행합니다. 루트 폴더에는 `START_HERE.md`, `requirements.txt`, `notebooks/`가 보입니다.


````python
from pathlib import Path

print("현재 폴더:", Path.cwd())
print("현재 폴더 파일 일부:")
for path in sorted(Path.cwd().iterdir())[:20]:
    print("-", path.name)
````

## uv와 Python의 관계

이번 과정에서는 Python 패키지 설치와 실행을 uv로 관리합니다.

```bash
uv --version
uv run --python 3.11 python --version
uv run --python 3.11 --with-requirements requirements.txt --with notebook jupyter notebook notebooks
```

전역 `python` 명령이 안 된다고 바로 python.org나 apt로 Python을 다시 설치하지 않습니다. 먼저 `uv run --python 3.11 python --version`이 되는지 확인합니다.


## sudo / apt가 필요한 경우

아래 같은 작업은 시스템 설정을 바꾸므로 스태프에게 먼저 요청합니다.

```bash
sudo apt update
sudo apt install ...
```

Python 패키지는 sudo/apt가 아니라 uv와 `requirements.txt`로 설치합니다.


## 결과 확인
